```python
import requests
from bs4 import BeautifulSoup
import random
import pandas as pd
import time

#选取的类别主要是：爱情、伤感、优美、想念、励志
#定义构造url和类别值的函数 接收类别和页面值
def get_type(order,list_num):
    types = ['【爱情语句】','【伤感语句】','【优美语句】','【想念的句子】','【励志的话】']
    urls = {
        '【爱情语句】':'https://www.duanwenxue.com/yuju/aiqing/',
        '【伤感语句】':'https://www.duanwenxue.com/yuju/shanggan/',
        '【优美语句】':'https://www.duanwenxue.com/yuju/youmei/',
        '【想念的句子】':'https://www.duanwenxue.com/yuju/xiangnian/',
        '【励志的话】':'https://www.duanwenxue.com/huayu/lizhi/'
        }
    #根据得到的类别号从字典中挑出主url进行url的构造
    type = types[int(order)-1]
    base_url = urls[type]
    url = base_url + "list_"+str(list_num)+".html"
    print("已经获取完毕"+type+"的第"+str(list_num)+"页，正在处理......")
    time.sleep(5)
    return type,url
    pass
```

```python
def get_onesen(url,type):
    startTime = time.time()
    re=requests.get(url)
    soup = BeautifulSoup(re.text,"html.parser")
    sentence_set=soup.find_all("div",{"class":"list-short-article"})[0].text
    t=sentence_set.split(str(type))
    sentence_list=[]
    for t_instence in t:
        if len(t_instence.split("\n\n")[0])>=6:
            sentence_list.append(t_instence.split("\n\n")[0])
    typerow = [type]*len(sentence_list)
    print("处理该页数据用时%.2f秒"%(time.time()-startTime))
    return sentence_list,typerow#返回该页的所有句子的一个列表
```

```python
#爬取某一类别的前200页的句子，放入数组存入csv
# 爱情：100######伤感：60########优美：60##励志：40
allsen = []
alltype = []
for i in range(1,40):
    sentype,url = get_type(5,i)
    senli,typerow = get_onesen(url,sentype)
    #获取一页之后停顿5s
    print("已经获取到第"+str(i)+"页")
    
    alltype = alltype+typerow
    allsen = allsen+senli
infos = {'type':alltype,'sen':allsen}
data = pd.DataFrame(infos, columns=['type', 'sen'])
data.to_csv(alltype[0] + ".csv",mode='a')
print("已经将内容写入文件保存")
```
